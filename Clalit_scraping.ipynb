### Cleaning the data

```python
# Some of the quistions have unrelated data regarding the author of the article, so we drop all the data from the first occurrence of the word ד"ר
def strip_from_dr(row):

  string = "".join(list(row['answer']))
  i = len(string)
  for i in range(len(string)-3):
    if string[i]== 'ד':
      if string[i+1]== '"':
        # print(string[i],string[i+1],string[i+2])
        # print(i)
        # print(string[:i:]) 
        print(string)
        break
  return string[:i:]

test = strip_from_dr(quistions_df.iloc[50])  
test 
```

```python
# Some of the quistions have unrelated data regarding the update date of the article, so we drop all the data from the first occurrence of the word עודכן
def strip_from_update(row):

  #print(list(row['answer']))
  string = "".join(list(row['answer']))
  i = len(string)
  for i in range(len(string)-5):
    if string[i]== 'ע' and string[i+1]== 'ו'and string[i+2]== 'ד' and string[i+3]== 'כ'and string[i+4]== 'ן':
        # print(string[i],string[i+1],string[i+2])
        # print(i)
        # print(string[:i:]) 
        print(string)
        break
  return string[:i:]
```

```python
quistions_df['answer'] = quistions_df.apply(strip_from_update, axis=1)
```

```python
quistions_df['answer'] = quistions_df.apply(strip_from_dr, axis=1)
```
