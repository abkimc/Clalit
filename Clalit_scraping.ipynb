### Cleaning the data

```python
# Some of the quistions have unrelated data regarding the author of the article, so we drop all the data from the first occurrence of the word ד"ר
def strip_from_dr(row):

  string = "".join(list(row['answer']))
  i = len(string)
  for i in range(len(string)-3):
    if string[i]== 'ד':
      if string[i+1]== '"':
        # print(string[i],string[i+1],string[i+2])
        # print(i)
        # print(string[:i:]) 
        print(string)
        break
  return string[:i:]

test = strip_from_dr(quistions_df.iloc[50])  
test 
```

```python
# Some of the quistions have unrelated data regarding the update date of the article, so we drop all the data from the first occurrence of the word עודכן
def strip_from_update(row):

  #print(list(row['answer']))
  string = "".join(list(row['answer']))
  i = len(string)
  for i in range(len(string)-5):
    if string[i]== 'ע' and string[i+1]== 'ו'and string[i+2]== 'ד' and string[i+3]== 'כ'and string[i+4]== 'ן':
        # print(string[i],string[i+1],string[i+2])
        # print(i)
        # print(string[:i:]) 
        print(string)
        break
  return string[:i:]
```

```python
quistions_df['answer'] = quistions_df.apply(strip_from_update, axis=1)
```

```python
quistions_df['answer'] = quistions_df.apply(strip_from_dr, axis=1)
```

### preparing the data for finetuning!

```python
sep = "</p>"
```

```python
# lets check that all questions ends with a ?
df['question']
target = len(df['question'])
sum = 0 
for i in df['question']:
  if (i[-1] == "?"):
    sum += 1

print("target: ", target," ,", " sum: ", sum  )
```

```python
# lets check that all questions contains the subject or a part of it

def Check_subject_in_quistion(row):
  
  subject = row['subject']
  question = row['question']

  # lets check how many words from the subject are in the quistion
  counter = 0 

  for word in row['subject'].split(" "):
    if word in question:
      counter += 1 

  return counter/len(row['subject'].split(" "))
```

```python
Check_subject_in_quistion(df.iloc[1])
```

```python
df["quistion_similarty_to_the_subject"] = df.apply(Check_subject_in_quistion, axis = 1 )
```

```python
df[df["quistion_similarty_to_the_subject"] <= 0.4]
```

```python
def Check_missing_subject_in_quistion(row):
  
  subject = row['subject']
  question = row['question']

  # let's check wich words from the subject are not in the answer
  missing = []

  for word in row['subject'].split(" "):
    if word not in question:
      missing.append(word)

  return missing
```

```python
Check_missing_subject_in_quistion(df.iloc[2])
```

```python
df["quistions_missing_words"] = df.apply(Check_missing_subject_in_quistion, axis = 1 )
```

```python
# geting the amount of missing words in every answer  
df["missing_words"] = df["quistions_missing_words"].apply(len)
```

```python
# Selecting only high similarity between qustion and answer 
quialty_df  = df[df["quistion_similarty_to_the_subject"] >= 0.5]
quialty_df
```
